```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

```python
df.index.name = 'job_index'
```

```python
df.head()
```

```python
df.index.dtype
```

```python
#reset_index()

df_usa = df[df['job_country'] == 'United States']

df_usa.reset_index(inplace=True)

df_usa
```

```python
#set_index()

df_usa.set_index('job_index', inplace=True)

df_usa
```

```python
#sort_index()

median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','min','max'])

median_pivot
```

```python
median_pivot.sort_values(by= ('median', 'salary_year_avg'), ascending=False, inplace=True)

median_pivot
```

```python
median_pivot.sort_index(inplace=True)
```
